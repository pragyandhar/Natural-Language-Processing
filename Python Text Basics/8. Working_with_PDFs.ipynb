```python
import PyPDF2 as pdf
```

```python
myfile = open('US_Declaration.pdf', mode='rb')
```

```python
pdf_reader = pdf.PdfReader(myfile) #always use PdfReader to read the file instead of PdfFileReader
```

```python
pdf_reader = pdf.PdfReader(myfile) # create a pdf reader object
```

```python
len(pdf_reader.pages) # get the number of pages in the pdf and reader.numPages is deprecated and was removed in PyPDF2 3.0.0. Use len(reader.pages) instead
```

```python
page_one = pdf_reader.pages[0] # get the first page of the pdf
```

```python
print(page_one.extract_text()) # extract the text from the first page
```

```python
myfile.close() # close the file
```

```python
f = open('US_Declaration.pdf', 'rb')
```

```python
pdf_reader = pdf.PdfReader(f)
```

```python
first_page = pdf_reader.pages[0]
```

```python
pdf_writer = pdf.PdfWriter()
```

```python
pdf_writer.add_page(first_page) # add the first page to the writer object and addPage is deprecated and was removed in PyPDF2 3.0.0. Use add_page instead.
```

```python
pdf_output = open('My_Brand_New.pdf', 'wb')
```

```python
pdf_writer.write(pdf_output) # write the new pdf to the file
```
